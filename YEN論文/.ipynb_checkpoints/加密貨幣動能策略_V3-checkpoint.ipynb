```python
for i in range(6,len(reward),1):
    rd = reward.iloc[i,30]
    if reward.iloc[i,30] == 0:
        print("熊式-----------------------------------")
        print("執行第",i) 
        q = reward.iloc[i:i+1,:30]  #取BBI值
#         print(q)
        a = q.T                    
#         print(a)
#         print("執行第",i)    
        c = a.sort_values(by=[i])
#         print(c)
        d = c.T

        e = d.iloc[:,0:5]
#         print(e)
        f= list(e.columns)
#         print(f)
        g = data.reindex(columns=f)
        h = g.shift(-7) / g.shift(-13) - 1 
        i = h.iloc[i-6,:]
        print(i)
        p = i.sum()
        result2 = pd.DataFrame(np.array([i]),
                   columns=f)
        result2.to_csv('result2.csv', mode='a', header=True)
        
    else :
        print("牛式===============================")
        print("執行第",i)     
        q = reward.iloc[i:i+1,:30] 
        a = q.T 
        c = a.sort_values(by=[i]) 
        d = c.T 
        e = d.iloc[:,25:30] 
        f= list(e.columns)
#         print(f)
        g = data.reindex(columns=f)
#         print(g)
        h = g.shift(-13)  / g.shift(-7) - 1 
#         print(h.head(25))
        i = h.iloc[i-6,:]
        print(i)
        p = i.sum()
        result2 = pd.DataFrame(np.array([i]),
                   columns=f)
        result2.to_csv('result2.csv', mode='a', header=True)
```
