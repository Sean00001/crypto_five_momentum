```python
data2 = pd.read_csv("全數據手動整理.csv")
```

```python
data = data.dropna()
```

```python
data['BTC'] = data['BTC'].str.replace(",","")
data['BTC'] = data['BTC'].astype(float)
data['ETH'] = data['ETH'].str.replace(",","") 
data['ETH'] = data['ETH'].astype(float)
data['BCH'] = data['BCH'].str.replace(",","") 
data['BCH'] = data['BCH'].astype(float)
```

```python
data['BTC']
seven = data.shift(6)
```

```python
today = data.drop(['Date','BTC','trend'], axis=1)
```

```python
sevenago = seven.drop(['Date','BTC','trend'], axis=1)
```

```python
sevenago.head(10)
```

```python
today = today.astype('float64')
```

```python
today
```

```python
today = today.astype(float, errors = 'raise')
```

```python
reward = today/sevenago-1  #計算觀察期
```

```python
reward["BBI"]= data2['trend'] #將牛式與熊式用BBI判斷寫入BBI
```

```python
reward.head(8)
```

```python
# reward.isnull().any()
```

```python
reward = reward.dropna(axis=0,how='any')   #將NA值拿掉
```

```python
reward
```

```python
reward.reset_index(drop=True, inplace=True)
reward
reward.to_csv('reward.csv', mode='a', header=True)
```

```python

for i in range(6,len(reward),1):
    num = i
    rd = reward.iloc[i,20]
    if reward.iloc[i,20] == 0:
        print("熊式-----------------------------------")
        print("執行第",i) 
        q = reward.iloc[i:i+1,:20]  #取BBI值
#         print(q)
        a = q.T               #將資料行轉列                 
#         print(a)
#         print("執行第",i)    
        c = a.sort_values(by=[i]) #將資料排序
        print(c)
        d = c.T            #將資料行轉列

        e = d.iloc[:,0:5]    #取後五筆資料
#         print(e)
        f= list(e.columns)    #將資料存進串列
#         print(f)
        g = data.reindex(columns=f)   #更改資料欄未名稱
        print("g================================================================", g.shift(-13)  )
        h = g / g.shift(7) - 1  #計算報酬
        
        i = h.iloc[i-6,:]
        print(i)
        p = i.sum()/5
        print("報酬",p,"%")
        result3 = pd.DataFrame(np.array([i]),
                   columns=f)
        result3.to_csv('all.csv', mode='a', header=True)
        print("i=",i)
        date = data.iloc[num,0]
#         date = str(date)
        datee = pd.Series(date)
        datee.to_csv('date.csv', mode='a', header=True)
    else :
        print("牛式===============================")
        print("執行第",i)     
        q = reward.iloc[i:i+1,:20] 
        a = q.T 
        c = a.sort_values(by=[i]) 
        print(c)
        d = c.T 
        e = d.iloc[:,15:20] 
        f= list(e.columns)
#         print(f)
        g = data.reindex(columns=f)
        print("g================================================================", g.shift(13)  )
        h = g.shift(-13)  / g.shift(-7) - 1 

#         print(h.head(25))
        i = h.iloc[i-6,:]
        print(i)
        p = i.sum()/5
        print("報酬",p,"%")
        result2 = pd.DataFrame(np.array([i]),
                   columns=f)
        result2.to_csv('all.csv', mode='a', header=True)
        print("i=",i)

        date = data.iloc[num,0]
#         date = str(date)
        datee = pd.Series(date)
        datee.to_csv('date.csv', mode='a', header=True)
```
