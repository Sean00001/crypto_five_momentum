## 偏態 與 峰態

```python
from scipy import stats
```

```python
## 偏態
skew = stats.skew(target_data['收盤價(元)'])
## 峰態 fisher=False 是 Eviews 作法
kurtosis = stats.kurtosis(target_data['收盤價(元)'], fisher=False)

print('skew      :  {}'.format(skew))
print('kurtosis  :  {}'.format(kurtosis))
```
