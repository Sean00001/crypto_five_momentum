```python
import pandas as pd
import numpy as np
from functools import partial, reduce
```

```python
def date_manipulate(x):
    x = x.str.strip()
    x = x.str.replace(' ', '')
    x = x.str.replace('年', '-')
    x = x.str.replace('月', '-')
    x = x.str.replace('日', '')
    x = x.str.replace('号', '')
    x = x.str.replace('.', '-')
    x = x.str.replace('—', '-')
    x = pd.to_datetime(x, errors='coerce')
    return x
```

```python
test = ['BTC','BTS','ADA']
data =['BTC','BTS','ADA']
for i in range(0,len(test)):
    data[i]= pd.read_csv(f'{test[i]}.csv',encoding="utf-8")
#     print(test[i])
    a = str(test[i])
#     print(a)
    data[i] = data[i].rename(columns={'收市': a,'日期':'Date'})
#     print("---------------------------------------------------------------")
#     print(test[i])
#     print(data[i].head(100))
    data[i] = data[i].drop(['開市', '高', '低', '成交量','更改%'], axis=1)
#     print("=====================================================================")
#     print(data[i])
   

df_final = data[0]
for i in range(1,len(data)):
#     df = str(df)
#     print(data[i])
    df_final = df_final.merge(data[i], on=['Date'], how='outer')
#     print(data[0])
df_final
# print(df_final)
df_final['Date'] = date_manipulate(df_final['Date'])
df_final = df_final.sort_values(by=['Date'])
df_final = df_final.dropna(axis=0,how='any')
df_final
```
